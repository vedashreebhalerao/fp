```python
df = df[['tenure' , 'age' , 'address' , 'income' , 'ed' , 'employ' , 'equip' , 'callcard' , 'wireless' , 'churn']]
df['churn'] = df['churn'].astype('int')
df.head()
```

```python
X = np.asarray(df[['tenure' , 'age' , 'address' , 'income' , 'ed' , 'employ' , 'equip' , 'callcard' , 'wireless' , 'churn']])
X[0:5]
```

```python
y = np.asarray(df['churn'])
y[0:5]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
Xtrain , Xtest , ytrain , ytest = train_test_split(X , y , test_size = 0.2 , random_state = 4)
print("Training Set : " , Xtrain.shape , ytrain.shape)
print("Testing Set : " , Xtest.shape , ytest.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C = 0.01 , solver = 'liblinear').fit(Xtrain , ytrain)
LR
```

```python
yhat = LR.predict(Xtest)
yhat
```

```python
yhat_prob = LR.predict_proba(Xtest)
yhat_prob
```
